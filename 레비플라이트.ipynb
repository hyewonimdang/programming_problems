```python
# 레비플라이트 1-3
import numpy as np
import matplotlib.pyplot as plt

def uniform_step(low_limit = 0, high_limit = 0.1):
    direction = np.random.choice((-1,1))
    move_L = np.random.uniform(low_limit, high_limit)
    return direction, move_L

class moving_agent():
    def __init__(self,initial = 0.01):
        self.X = initial
        self.Distance = 0
        self.Trial = 0
    def stepping(self):
        direction, move_L = uniform_step()
        self.X += direction*move_L
        self.Trial += 1
        self.Distance += move_L
    def check(self):
        Done = False
        if self.X < 0:
            Done = True
            self.Distance += self.X
        elif self.X >1:
            Done = True
            self.Distance -= (self.X-1)
        else:
            pass
        return Done
    
N_Trials = 10000

distance_set = []
distance_mean_set = []
Trial_set = []

for i in range(N_Trials):
    agent = moving_agent()
    Done = False
    while not Done:
        agent.stepping()
        Done = agent.check()
    distance_set.append(agent.Distance)
    distance_mean_set.append(np.mean(Distance_set))
    Trial_set.append(agent.Trial)
print('평균 움직인 거리:', np.mean(Distance_set))
print('횟수:', np.mean(Trial_set))
```

```python
import numpy as np
import random

def initialize_grid(rows, cols):
    return np.random.choice([-1, 1], size=(rows, cols))

def count_arrows_around(grid, x, y):
    rows, cols = grid.shape
    neighbors = [(x+i, y+j) for i in range(-1, 2) for j in range(-1, 2) if (i, j) != (0, 0)]
    arrow_counts = [grid[i % rows, j % cols] for i, j in neighbors]
    return arrow_counts.count(1), arrow_counts.count(-1)

def apply_ising_model(grid):
    rows, cols = grid.shape
    new_grid = grid.copy()

    for i in range(rows):
        for j in range(cols):
            up_arrows, down_arrows = count_arrows_around(grid, i, j)

            if up_arrows > down_arrows and random.random() < 2/3:
                new_grid[i, j] = 1
            elif down_arrows > up_arrows and random.random() < 2/3:
                new_grid[i, j] = -1

    return new_grid

def print_grid(grid):
    for row in grid:
        print(' '.join(['↑' if arrow == 1 else '↓' for arrow in row]))

def main():
    rows = int(input("Enter the number of rows: "))
    cols = int(input("Enter the number of columns: "))
    steps = int(input("Enter the number of simulation steps: "))

    grid = initialize_grid(rows, cols)
    print("Initial Configuration:")
    print_grid(grid)

    for step in range(steps):
        grid = apply_ising_model(grid)

    print("\nFinal Configuration:")
    print_grid(grid)

if __name__ == "__main__":
    main()
```
